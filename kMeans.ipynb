```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('/content/fool.jpg')

if img is None:
    print("Error: Unable to load image from the specified path.")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.title("Original Image")
```

```python
import numpy as np
img.shape
pixels = img.reshape((-1, 3))
print("Reshaped image shape:", pixels.shape)
```

```python
from sklearn.cluster import KMeans
K =
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(pixels)
new_colors = kmeans.cluster_centers_[kmeans.labels_]
compressed_img = new_colors.reshape(img.shape).astype('uint8')
```

```python
plt.imshow(compressed_img)
plt.axis('off')
plt.title(f"Compressed Image with {K} colors")
plt.show()
```
